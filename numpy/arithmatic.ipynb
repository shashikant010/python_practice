```python
array1=np.array([1,2,3,4])
```

```python
print("the minimum value in the array is : ",np.min(array1),"at",np.argmin(array1))
print("the maximum value in the array is : ",np.max(array1),"at",np.argmax(array1))
```

```python
array2 = np.array([[1,2,3,4],[5,6,7,8]])
```

```python
print("the minimum value along row in array is : ",np.min(array2,axis=0))
print("the minimum value along column in array is : ",np.min(array2,axis=1))
```
