Data types

```python
integerArray = [1,2,3,4]
print(type(integerArray))
```

```python
import numpy as np
numpyArray = np.array([1,2,3,4])
print(type(numpyArray))
```

```python
numpyFloatArray = np.array([1.0,2.0,3.0,4.0])
print(numpyFloatArray.dtype)
```

```python
numpyStringArray = np.array(["1","2"])
print(numpyStringArray.dtype)
```

```python
numpyMixArray = np.array(["1","2",2])
print(numpyMixArray.dtype)
```

```python
x =  np.array([1,2,3,4],dtype="int32")
print(x.dtype)
```

```python
y =  np.array([1,2,3,4],dtype="float")
print(y)
print(y.dtype)
```
