```python
import pandas as pd
```

```python
dataframe1=pd.DataFrame({"A":[1,2,3,4],"B":[22,11,55,44]})
dataframe2=pd.DataFrame({"A":[1,2,3,4],"C":[65,43,23,63]})
pd.merge(dataframe1,dataframe2,on="A")
```

```python
dataframe3=pd.DataFrame({"A":[1,2,3,4],"B":[22,11,55,44]})
dataframe4=pd.DataFrame({"A":[1,2,3,5],"C":[65,43,23,63]})
pd.merge(dataframe3,dataframe4,on="A")
```

```python
pd.merge(dataframe3,dataframe4,on="A",how="inner")
```

```python
pd.merge(dataframe3,dataframe4,on="A",how="left")
```

```python
pd.merge(dataframe3,dataframe4,on="A",how="right")
```

```python
pd.merge(dataframe3,dataframe4,on="A",how="outer",indicator=True)
```

```python
dataframe5=pd.DataFrame({"A":[1,2,3,4],"B":[22,11,55,44]})
dataframe6=pd.DataFrame({"A":[1,2,3,5],"B":[65,43,23,63]})
pd.merge(dataframe5,dataframe6)
```

```python
pd.merge(dataframe5,dataframe6,left_index=True,right_index=True)
```

```python
pd.merge(dataframe5,dataframe6,left_index=True,right_index=True,suffixes=["sky","shashi"])
```

```python
pd.concat([dataframe1,dataframe2])
```

```python
pd.concat([dataframe1,dataframe2],axis=1)
```

```python
pd.concat([dataframe1,dataframe2],join="inner") #outer is default
```

```python
pd.concat([dataframe1,dataframe2],keys=["data1","data2"])
```
