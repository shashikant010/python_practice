We can add axis parameter in dropna method : axis=0 remove data in row and axix=1 removes data in column

in dropna how="any" will drops all rows with na  and how="all" will remove only that row which has all null value

in dropna subset=["columnName","columnName"] can be used to detect and remove null values along that particular column

inplace=True can be use to inplace the value to other variable

thresh=count will drop the column with count number of null values other wise will not drop it

csv1.fillna(method="ffill") : will fill the null with next non-null value available
